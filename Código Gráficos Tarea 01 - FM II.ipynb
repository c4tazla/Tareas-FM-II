```python
import numpy as np
import matplotlib.pyplot as plt

def f(t, a):
    return np.where(np.abs(t) < 1/a, 1 - a*np.abs(t), 0)

def F(w, a):
    return (1/a) * (np.sinc(w/(2*np.pi*a)))**2

t = np.linspace(-5, 5, 1000)
w = np.linspace(-20, 20, 1000)

a_values = [0.5, 1, 2]

for a in a_values:
    plt.figure()
    plt.plot(t, f(t, a))
    plt.title(f"f(t) para a = {a}")
    plt.xlabel("t")
    plt.ylabel("f(t)")
    plt.grid()
    plt.show()

    plt.figure()
    plt.plot(w, F(w, a))
    plt.title(r"Transformada de Fourier F[f(t)]($\omega$) para a = " f"{a}")
    plt.xlabel(r"$\omega$")
    plt.ylabel(r"F[f(t)]($\omega$)")
    plt.grid()
    plt.show()
```
